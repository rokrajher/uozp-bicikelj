### Priprava testne in trening množice

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, Normalizer

# create X and y sets by selecting the appropriate columns from the DataFrame
X = df.iloc[:, 83:]
y = df.iloc[:, :83]

display(X.tail(5))
display(y.tail(5))

# STANDARDIZACIJA

# NORMALIZACIJA

# split data into 70% training set and 30% test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Training set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)
```

#### Funkcija za izvedbo prečnega preverjanja

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate
from scipy import stats

def cross_validation(model, X_train, y_train, n_folds=5):
# Create an instance of Leave-One-Out Cross-Validation
    cv = KFold(n_splits=n_folds)

    # Calculate the cross-validation score for your model
    scoring = ['neg_mean_squared_error', 'neg_mean_absolute_error']#, 'r2']
    scores = cross_validate(model, X_train, y_train, scoring=scoring, cv=cv, n_jobs=-1)

    return scores
```

#### Funkcija za evaluacijo kvalitete modelov

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def display_scores(y_true, y_pred):
    mse = mean_squared_error(y_true,y_pred).round(3)
    mae = mean_absolute_error(y_true,y_pred).round(3)
    rmse  = np.sqrt(mse).round(3)
    r2 = r2_score(y_true,y_pred).round(3)

    data = np.array([[mse,mae,rmse,r2]])
    scores = pd.DataFrame(data, columns=['MSE','MAE','RMSE','R-squared'])

    display(scores)
    
    return data
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def display_cv_scores(scores, n_folds=5):
    
    # Calculate the mean and standard deviation of the cross-validation scores
    # Convert the negative MSE and MAE scores to positive values
    mse_scores = -scores['test_neg_mean_squared_error']
    mae_scores = -scores['test_neg_mean_absolute_error']
    #r2_scores = scores['test_r2']


    # # Extract the R-squared scores
    # r2_scores = scores[2]

    # Calculate the mean and standard deviation of the scores for each metric
    mse_mean = np.mean(mse_scores).round(2)
    mse_std = np.std(mse_scores).round(2)
    mae_mean = np.mean(mae_scores).round(2)
    mae_std = np.std(mae_scores).round(2)
    # r2_mean = np.mean(r2_scores)
    # r2_std = np.std(r2_scores)
    t_value = stats.t.ppf(1-0.025, n_folds-1)  # t-value for 95% confidence interval

    ci_low = mae_mean - t_value * mae_std / np.sqrt(n_folds)
    ci_high = mae_mean + t_value * mae_std / np.sqrt(n_folds)

    # Print the mean and standard deviation of the cross-validation scores for each metric
    print(f"Mean MSE: {mse_mean:.2f}, Standard deviation: {mse_std:.2f}")
    print(f"Mean MAE: {mae_mean:.2f}, Standard deviation: {mae_std:.2f}")
    #print(f"Mean R-squared: {r2_mean:.2f}, Standard deviation: {r2_std:.2f}")
    print(f"95% Confidence interval for the mean: ({ci_low:.2f}, {ci_high:.2f})")
    
    data = np.array([[mse_mean, mse_std, mae_mean, mae_std]])
    scores_df = pd.DataFrame(data, columns=['MSE mean','MSE std', 'MAE mean','MAE std'])
    display(scores_df)
    
    return scores_df
```

##### Grid Search

```python
from sklearn.model_selection import GridSearchCV

def get_best_model(model, X_train, y_train, param_grid):
    # Perform a grid search to find the best parameters
    grid_search = GridSearchCV(model, param_grid=param_grid, cv=5, n_jobs=-1)
    grid_search.fit(X_train, y_train)

    # Return the best model
    best_model = grid_search.best_estimator_
    return best_model
```
