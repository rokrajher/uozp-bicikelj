```python
from sklearn.preprocessing import StandardScaler, Normalizer

def preprocess_data(df, suff):
    df_postaje = df.iloc[:, 1:83]

    # Convert 'timestamp' column to timestamptime
    df['timestamp'] = pd.to_datetime(df['timestamp'])

    # Extract day, hour, minute, and second values
    df['month'] = df['timestamp'].dt.month
    df['day'] = df['timestamp'].dt.day
    df['hour'] = df['timestamp'].dt.hour
    #df['minute'] = df['timestamp'].dt.minute
    df['dayofweek'] = df['timestamp'].dt.dayofweek
    df['rush_hour'] = ((df['hour'] >= 7) & (df['hour'] <= 9)) | ((df['hour'] >= 15) & (df['hour'] <= 17))
    df['is_weekend'] = (df['dayofweek'] == 5) | (df['hour'] == 6)
    # df['is_year_end'] = df['timestamp'].dt.is_year_end
    # df['is_month_start'] = df['timestamp'].dt.is_month_start
    # df['is_month_end'] = df['timestamp'].dt.is_month_end
    # df['is_quarter_start'] = df['timestamp'].dt.is_quarter_start
    # df['is_quarter_end'] = df['timestamp'].dt.is_quarter_end
    # create a list of bins for the pd.cut() function
    bins = [-1, 6, 9,15, 18, 23]

    # create a list of labels for the bins
    labels = [0,1,2,3,4]

    # use pd.cut() to categorize the hours into bins and assign the labels
    #df['perioud'] = pd.cut(df['hour'], bins=bins, labels=labels)

    #df = df.drop(columns=['hour'])
    df = df.drop(columns=['timestamp'])
    
    df_closest = pd.read_csv(f'./generiraj/closest_time_{suff}.csv')
    df_closest = df_closest.add_suffix('_closest')
    df = df.join(df_closest)
    df = df.drop(columns=['timestamp_closest'])

    #result = pd.merge(df, df_closest, on=['timestamp', ''])    
    
    return df

    
    
df = preprocess_data(df, "train")
display(df.tail(5))
```

#### Funkcija za izvedbo prečnega preverjanja

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate
from scipy import stats

def cross_validation(model, X_train, y_train, n_folds=5):
# Create an instance of Leave-One-Out Cross-Validation
    cv = KFold(n_splits=n_folds)

    # Calculate the cross-validation score for your model
    scoring = ['neg_mean_squared_error', 'neg_mean_absolute_error']#, 'r2']
    scores = cross_validate(model, X_train, y_train, scoring=scoring, cv=cv, n_jobs=-1)

    return scores
```

#### Funkcija za evaluacijo kvalitete modelov

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def display_scores(y_true, y_pred):
    mse = mean_squared_error(y_true,y_pred).round(4)
    mae = mean_absolute_error(y_true,y_pred).round(4)
    rmse  = np.sqrt(mse).round(4)
    r2 = r2_score(y_true,y_pred).round(4)

    data = np.array([[mse,mae,rmse,r2]])
    scores = pd.DataFrame(data, columns=['MSE','MAE','RMSE','R-squared'])

    display(scores)
    
    return data
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def display_cv_scores(scores, n_folds=5):
    
    # Calculate the mean and standard deviation of the cross-validation scores
    # Convert the negative MSE and MAE scores to positive values
    mse_scores = -scores['test_neg_mean_squared_error']
    mae_scores = -scores['test_neg_mean_absolute_error']
    #r2_scores = scores['test_r2']


    # # Extract the R-squared scores
    # r2_scores = scores[2]

    # Calculate the mean and standard deviation of the scores for each metric
    mse_mean = np.mean(mse_scores).round(4)
    mse_std = np.std(mse_scores).round(4)
    mae_mean = np.mean(mae_scores).round(4)
    mae_std = np.std(mae_scores).round(4)
    # r2_mean = np.mean(r2_scores)
    # r2_std = np.std(r2_scores)
    t_value = stats.t.ppf(1-0.025, n_folds-1)  # t-value for 95% confidence interval

    ci_low = mae_mean - t_value * mae_std / np.sqrt(n_folds)
    ci_high = mae_mean + t_value * mae_std / np.sqrt(n_folds)

    # Print the mean and standard deviation of the cross-validation scores for each metric
    print(f"Mean MSE: {mse_mean:.2f}, Standard deviation: {mse_std:.2f}")
    print(f"Mean MAE: {mae_mean:.2f}, Standard deviation: {mae_std:.2f}")
    #print(f"Mean R-squared: {r2_mean:.2f}, Standard deviation: {r2_std:.2f}")
    print(f"95% Confidence interval for the mean: ({ci_low:.2f}, {ci_high:.2f})")
    
    data = np.array([[mse_mean, mse_std, mae_mean, mae_std]])
    scores_df = pd.DataFrame(data, columns=['MSE mean','MSE std', 'MAE mean','MAE std'])
    display(scores_df)
    
    return scores_df
```

##### Grid Search

```python
from sklearn.model_selection import GridSearchCV

def get_best_model(model, X_train, y_train, param_grid):
    # Perform a grid search to find the best parameters
    grid_search = GridSearchCV(model, param_grid=param_grid, cv=5, n_jobs=-1)
    grid_search.fit(X_train, y_train)

    # Return the best model
    best_model = grid_search.best_estimator_
    return best_model
```
